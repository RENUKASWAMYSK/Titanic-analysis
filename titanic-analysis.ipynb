```python
import seaborn as sns 
import pandas as pd
data = sns.load_dataset('titanic')
data.head()
```

```python
print(data.isnull().sum())
```

```python
data['embarked'] = data['embarked'].fillna(data['embarked'].mode()[0])
```

```python

data= data.dropna()
```

```python
data.isnull().sum()
```

numpy is so fast compared to traditional python lists. 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.countplot(data = data,x = 'who',hue = 'pclass', palette = 'muted')
plt.title("Passenger Class Distribution")
plt.xlabel('class')
plt.ylabel('count')
```

.shape represents to check the size of the data first a was 3 elements present and in c was 2 elements within that 4 elements was present.

```python
survivors = data[data['survived']==1]
gender_count = survivors['sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_count, labels= gender_count.index, autopct = "%1.1f%%", colors =['yellow','blue'],startangle = 140)
plt.title("Gender Distribution in pie chart")
plt.axis('equal')
```

```python
fig, axes = plt.subplots(1,2, figsize=(5,5))
sns.countplot(data = data, x='pclass',hue='sex',ax = axes[0],palette='pastel')
axes[0].set_title('Passenger Class Distribution')
axes[1].pie(gender_count, labels = gender_count.index, autopct = '%1.1f%%', colors=['yellow','blue'], startangle=140)
axes[1].set_title('Gender Distribution Survivors')
axes[1].axis('equal')

plt.tight_layout()
```
